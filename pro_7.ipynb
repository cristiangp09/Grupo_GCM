```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/labeconometria/proyectos/main/economicas.csv")
```

```python
df.columns=[col.lower() for col in df.columns]
```

```python
df.drop(columns=['result_nombreprueba','inst_nombre_institucion','estu_estadoinvestigacion','inst_cod_institucion','estu_tipodocumento','estu_pagomatriculapropio','estu_agregado','estu_estadoinvestigacion','inst_caracter_academico','estu_depto_presentacion','estu_mcpio_presentacion','inst_origen','estu_nucleo_pregrado','estu_paisdocumentos11','estu_otrocole_termino'], inplace=True)
```

```python
df['estu_exterior'].replace({'SI':1, "NO":0}, inplace=True)
```

```python
df.dropna(inplace=True) #sabemos que eliminar las filas que contengan missing values pero dado que es cuestion de tiempo y por practicidad del ejercicio se tomo esta decision
```

```python
df.info()
```

#Visualización univariada de los datos y estadistica descritiva

```python
df['punt_global'].describe() #Estadistica descriptiva de los puntajes globales de la prueba Saber Pro
```
