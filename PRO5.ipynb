#**Outliers :0**

```python
df2=df[["price","minimum_nights","number_of_reviews","reviews_per_month","calculated_host_listings_count","availability_365"]]
```

```python
from scipy.spatial.distance import mahalanobis
import numpy as np
from scipy.stats import chi2
import scipy.stats as st
```

```python
mahal_distances = []

for row in df2.to_numpy():
  mahal_distances.append(mahalanobis(row, df2.mean(), np.linalg.inv(np.cov(df2.values.T))))
```

```python
df['mahal_distances'] = mahal_distances
```

```python
df['p_value'] = 1 - st.chi2.cdf(df['mahal_distances'], 13-1)
```

```python
df #New df with mahal_distances and p_value columns
```

We decided to filter all the data

```python
df = df[df['p_value'] > 0.01]
```
